```python
from scipy.fftpack import fft, ifft
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of samples
N = 1024
n= np.arange(N)
x = np.linspace(0, 2.0 * np.pi,N)
f  = x

fig, axes = plt.subplots(1,4,figsize=(15,2))
ax1 = axes[0]
ax2 = axes[1]
ax3 = axes[2]
ax4 = axes[3]

# Sine function
y = [np.sin(i) for i in f] 
ax1.plot(f,  (y))

#close up
ax2.plot(f,(y), lw=2.0, c='b')
ax2.set_xlim(0,5)
ax2.set_title('close-up', fontsize= 16, fontweight="bold")
ax2.set_ylabel('FFT magnitude (power) - log')
ax2.set_xlabel('Frequency (Hz)')

#fourior spectrum

fft1 = np.fft.fft(y)
z = np.fft.fftshift((fft1))
print(np.abs(z))
ax3.plot(np.fft.fftfreq(len(x)), np.abs(fft1) )
ax3.set_ylabel("Amplitude")
ax3.set_xlabel("Frequency")

#phase angle
p = np.angle(fft1)
ax4.plot(np.fft.fftshift(np.fft.fftfreq(len(x))), p)
plt.subplots_adjust(wspace=1.0)


plt.show()
```

```python

def fft2d(image):
   
    fftcols = np.array([fft(row) for row in image]).transpose()

    return np.array([fft(row) for row in fftcols]).transpose()


final = fft2d((img))
```

```python

# Calculate the magnitude and phase spectra
fftshift= np.fft.fftshift(final)

magnitude_spectrum = 20*np.log10(np.abs(fftshift))
phase_spectrum = np.angle(fftshift)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.imshow(phase_spectrum, cmap='gray')
plt.title('Phase spectrum'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
K = fftshift.copy()

# Circular HPF mask, center circle is 0, remaining all ones
rows, cols = img.shape
crow, ccol = int(rows / 2), int(cols / 2)

mask = np.ones((rows, cols), np.uint8)
r = 5
center = [crow, ccol]
x, y = np.ogrid[:rows, :cols]
mask_area = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= r*r
mask[mask_area] = 0

fshift = fftshift * mask


plt.imshow(np.log(1+ np.abs(fshift)), plt.cm.gray)

plt.show()
```

```python
f_ishift = np.fft.ifftshift(fshift)

ifft2 = np.fft.ifft2(f_ishift)

plt.imshow(np.log10(1+ np.abs(ifft2)), plt.cm.gray)

plt.show()
```
